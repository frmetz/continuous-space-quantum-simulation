```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
size = 14
plt.rc('font', family='serif')
plt.rc('text', usetex=True)
plt.rc('xtick', labelsize=size)
plt.rc('ytick', labelsize=size)
plt.rc('axes', labelsize=size)
plt.rc('legend', fontsize=size)
plt.rc('legend', handlelength=2)
plt.rc('font', size=size)
plt.rc('lines', markersize=8)
plt.rcParams['figure.dpi'] = 100

linestyles = ["-", ":", "-.", "--", (0, (3, 1, 1, 1, 1, 1)), "-", "--", "-.", ":"]
colors = ['#e4bf44', '#c65742', '#9ad0bb', '#87584E', '#aba18d', '#332737', '#e4bf44', '#c65742', '#9ad0bb', '#aba18d', '#332737']
markers = ["o", "v", "s", "d", "X", "p", "*", "o", "v", "s", "d", "*"]
```

```python
jastrow_ens = np.load("data/jastrow_energies.npy")
hybrid_ens = np.load("data/hybrid_energies.npy")
```

# Fig. 9

```python
fig = plt.figure(figsize =(6.4*1.85, 4.8*0.7))

colspan = 6
rowspan = 5
nr,nc = rowspan, 2*colspan +2

plt11 = plt.subplot2grid((nr,nc), (0, 0), rowspan=rowspan, colspan=colspan) 
plt12 = plt.subplot2grid((nr,nc), (0, colspan+2), rowspan=rowspan, colspan=colspan) 

ax  = [plt11, plt12]

##############################################
# Subplot (a)
##############################################
i=0
ax[i].axhline(y=-2.764882842300, color=colors[-1], linestyle='-.', label=r"Mean-field")
ax[i].axhline(y=-3.083834603828697, color=colors[-2], linestyle='--')

n1 = len(jastrow_ens)
x1 = np.linspace(1, n1, n1)

ax[i].plot(x1, np.array(jastrow_ens), label=r"Slater-Jastrow", color=colors[-2])

ax[i].set_xlabel("training iterations")
ax[i].set_ylabel(r"$E\, r_s$")
ax[i].set_ylim(-3.12, -2.72)
ax[i].legend(loc='center right',frameon=False)
ax[i].text(-0.2, 1.0, "(a)", transform=ax[i].transAxes, fontsize=size, verticalalignment='top')

##############################################
# Subplot (b)
##############################################
i=1
ax[i].axhline(y=-3.088341655547131, color=colors[1], linestyle=':')
ax[i].axhline(y=-3.083834603828697, color=colors[-2], linestyle='--')

n1 = len(jastrow_ens)
n2 = len(hybrid_ens)
x1 = np.linspace(1, n1, n1)
x2 = np.linspace(n1+1, n1+1+n2, n2)

ax[i].plot(x2, hybrid_ens, label=r"PQC Backflow", color=colors[1])

ax[i].set_xlabel("training iterations")
ax[i].set_ylabel(r"$E\, r_s$")
ax[i].legend(frameon=False)
ax[i].text(-0.2, 1.0, "(b)", transform=ax[i].transAxes, fontsize=size, verticalalignment='top')

# plt.savefig(f'_heg.png', dpi=300,bbox_inches='tight')
# plt.savefig(f'_heg.pdf', dpi=300,bbox_inches='tight')
```
